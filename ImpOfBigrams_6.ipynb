```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from collections import Counter
nltk.download('punkt_tab')

corpus = """
Natural language processing (NLP) is a subfield of artificial intelligence
that focuses on the interaction between computers and humans through natural language.
The goal of NLP is to enable computers to understand, interpret, and generate human language.
It involves tasks such as part-of-speech tagging, named entity recognition,
language generation, and more.
"""


tokens = word_tokenize(corpus.lower())

unigrams = list(tokens)
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))
fourgrams = list(ngrams(tokens, 4))

unigram_counter = Counter(tokens)
bigram_counter = Counter(ngrams(tokens, 2))
trigram_counter = Counter(ngrams(tokens, 3))
fourgram_counter = Counter(ngrams(tokens, 4))


total_unigrams = sum(unigram_counter.values())
total_bigrams = sum(bigram_counter.values())
total_trigrams = sum(trigram_counter.values())
total_fourgrams = sum(fourgram_counter.values())


print("Unigrams:", unigrams)
print("\nBigrams:", bigrams)
print("\nTrigrams:", trigrams)
print("\n4-grams:", fourgrams)

def calculate_ngram_probability(sentence, ngram_counter, total_ngrams, n):
    sentence_tokens = word_tokenize(sentence.lower())
    sentence_ngrams = list(ngrams(sentence_tokens, n))
    probability = 1.0

    for ngram in sentence_ngrams:
        count_ngram = ngram_counter[ngram]
        probability *= count_ngram / total_ngrams if total_ngrams > 0 else 0

    return probability


sentence = "natural language processing"


unigram_prob = calculate_ngram_probability(sentence, unigram_counter, total_unigrams, 1)
bigram_prob = calculate_ngram_probability(sentence, bigram_counter, total_bigrams, 2)
trigram_prob = calculate_ngram_probability(sentence, trigram_counter, total_trigrams, 3)
fourgram_prob = calculate_ngram_probability(sentence, fourgram_counter, total_fourgrams, 4)

print(f"Unigram Probability: {unigram_prob:.6e}")
print(f"Bigram Probability: {bigram_prob:.6e}")
print(f"Trigram Probability: {trigram_prob:.6e}")
print(f"4-gram Probability: {fourgram_prob:.6e}")
```
