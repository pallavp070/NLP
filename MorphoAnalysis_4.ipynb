```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python

lemmatizer = WordNetLemmatizer()

sentence = "Unhappily, the miscommunication between overconfident managers and inexperienced workers led to disorganized workflows and underperformance. The reexamination of the process proved to be inefficient and unnecessarily complicated. With continuous misunderstanding, the restructured plan became unmanageable, and employees felt increasingly dissatisfied with their assignments. Overemphasis on perfectionism created a disproportionate imbalance in task allocation, leading to unexpected underachievement. In the end, the restructuring caused a complete disintegration of team dynamics, making it nearly impossible to reestablish productivity."

words = word_tokenize(sentence)


pos_tags = pos_tag(words)

print("Original words and their POS tags:")
for word, tag in pos_tags:
    print(f"{word}: {tag}")

print("\nMorphological analysis and POS tags after lemmatization:")
```

```python
for word, tag in pos_tags:

    if tag.startswith('VB'):
        lemma = lemmatizer.lemmatize(word, pos='v')
    elif tag.startswith('JJ'):
        lemma = lemmatizer.lemmatize(word, pos='a')
    elif tag.startswith('NN'):
        lemma = lemmatizer.lemmatize(word, pos='n')
    else:
        lemma = word

    print(f"Before: {word} (POS: {tag}) --> After: {lemma} (POS: {tag})")
```
