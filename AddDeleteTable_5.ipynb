```python
import pandas as pd

# Define the Add-Delete table with examples of transformations
add_delete_table = {
    'ed': {'add': ['walk', 'dance', 'jump'], 'delete': ['run', 'fly', 'swim'], 'irregular': {'run': 'ran', 'fly': 'flew', 'swim': 'swam'}},
    'ing': {'add': ['play', 'sing', 'teach'], 'delete': ['eat'], 'irregular': {'eat': 'eating'}},
    's': {'add': ['dog', 'book', 'cat'], 'delete': []}
}

# Function to process word transformation
def process_word(word, rule):
    if rule in add_delete_table:
        action = add_delete_table[rule]

        # Handle irregular transformations
        if word in action.get('irregular', {}):
            return action['irregular'][word]

        # Add operation (adding suffixes)
        if word in action['add']:
            return word + rule

        # Delete operation (removing characters or applying irregular forms)
        elif word in action['delete']:
            return word[:-1]  # Example of deleting the last character
    return word  # If no transformation is applicable, return original word

# Create a list of example words to apply transformations
example_words = ["walk", "play", "dog", "run", "eat", "jump", "dance", "fly", "teach", "sing"]

# Create a corresponding rules list (for each word, we pick a rule to apply)
rules = ['ed', 'ing', 's', 'ed', 'ing', 'ing', 'ed', 'ed', 'ing', 'ing']

# Apply the transformations and store results
outputs = [process_word(word, rule) for word, rule in zip(example_words, rules)]

# Create the DataFrame to display the table
data = {
    "Word": example_words,
    "Rule": rules,
    "Operation": ["Add" if rule != "ed" else "Delete (Irregular)" for rule in rules],  # Simplified Operation column
    "Output": outputs
}

df = pd.DataFrame(data)

# Display the resulting table
df
```
